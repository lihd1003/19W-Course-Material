```python
import numpy as np
import matplotlib.pyplot as plt
def fx(x):
    return x**2 - 3*x

x = np.linspace(-10, 10, 100)
f = fx(x)
ff = fx(f)
fff = fx(ff)
plt.plot(x, f)
plt.plot(f, ff)
plt.plot(ff, fff)
```

```python
t = np.linspace(0.001, 1, 1000)
s = 1
test = (t**4 + t**2) / (t**4 + 2 * t**2 + 1) 
fitted  = (s**4 + t**2) / (s**4 + 2 * s**2 + 1) 
plt.plot(t, test, label="weights from MLE of test dataset")
plt.ylabel("loss")
plt.xlabel("sigma test")
plt.plot(t, fitted, label="weights from MLE of train dataset")
plt.title("Loss on the test dataset using different fitted weight")
plt.legend();
```
